```python
#import pkgs
import numpy as np
import pandas as pd
from scipy import stats
from statistics import mean
```

```python
#import f1 and fpr results from models
%store -r f1_simple_case1
%store -r fpr_simple_case1

%store -r f1_simple_case2
%store -r fpr_simple_case2

%store -r f1_simple_case3
%store -r fpr_simple_case3

#store descriptive statistics for each case
score_simple = [mean(f1_simple_case1), np.std(f1_simple_case1), mean(fpr_simple_case1), np.std(fpr_simple_case1), mean(f1_simple_case2), np.std(f1_simple_case2), mean(fpr_simple_case2), np.std(fpr_simple_case2), mean(f1_simple_case3), np.std(f1_simple_case3), mean(fpr_simple_case3), np.std(fpr_simple_case3)]

#round statistics
for i in range(len(score_simple)):
    score_simple[i] = round(score_simple[i], 4)
    
#check list
score_simple
```

```python
#run kruskal-wallis test
res_simple = stats.kruskal(f1_simple_case1, f1_simple_case2, f1_simple_case3)
res_simple_fpr = stats.kruskal(fpr_simple_case1, fpr_simple_case2, fpr_simple_case3)

#store both results in one list
res_simple_comb = list(res_simple) + list(res_simple_fpr)
```
